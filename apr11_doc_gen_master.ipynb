````python
## 1. Extract content from a DOCX file, capturing headers and their associated content.

def extract_document_content(doc_path: str) -> List[Dict[str, Any]]:
    """
    Extract content from a DOCX file, capturing headers and their associated content.
    
    Args:
        doc_path: Path to the DOCX file
        
    Returns:
        List of dictionaries containing header and content information
    """
    doc = Document(doc_path)
    sections = []
    current_header = None
    current_content = []
    
    for paragraph in doc.paragraphs:
        if paragraph.style.name.startswith('Heading'):
            # If we already have a header being processed, save it
            if current_header:
                sections.append({
                    'header': current_header,
                    'content': '\n'.join(current_content),
                    'level': int(current_header_level) if current_header_level else 1
                })
            
            # Start new header
            current_header = paragraph.text
            current_header_level = paragraph.style.name.replace('Heading', '')
            current_content = []
        else:
            # Add content to current header if there is text
            if paragraph.text.strip():
                current_content.append(paragraph.text)
    
    # Don't forget to add the last section
    if current_header:
        sections.append({
            'header': current_header,
            'content': '\n'.join(current_content),
            'level': int(current_header_level) if current_header_level else 1
        })
        
    return sections
````

````python
## 2. function for converting big document to small chunks in json format

def chunk_sections(sections: List[Dict[str, Any]], max_chunk_size: int = 5) -> List[List[Dict[str, Any]]]:
    """
    Split sections into chunks of approximately equal size.
    
    Args:
        sections: List of section dictionaries
        max_chunk_size: Maximum number of sections per chunk
        
    Returns:
        List of chunks, where each chunk is a list of section dictionaries
    """
    chunks = []
    current_chunk = []
    
    for section in sections:
        current_chunk.append(section)
        
        # Create a new chunk if the current one reaches max size
        if len(current_chunk) >= max_chunk_size:
            chunks.append(current_chunk)
            current_chunk = []
    
    # Add the last chunk if it has any content
    if current_chunk:
        chunks.append(current_chunk)
        
    return chunks
````

````python
## 4. Create a prompt for the LLM to reformat the content
## option b

def create_prompt(chunk: List[Dict[str, Any]], template_example: str) -> str:
    """
    Create a prompt for the LLM to reformat the content.
    
    Args:
        chunk: A list of section dictionaries
        template_example: Example of the new template format
        
    Returns:
        A formatted prompt string
    """
    prompt = f"""
You are tasked with reformatting document content to match a new template style.

NEW TEMPLATE FORMAT EXAMPLE:
{template_example}

ORIGINAL CONTENT TO REFORMAT:
"""
    
    for section in chunk:
        prompt += f"\n## HEADING (Level {section['level']}): {section['header']}\n"
        prompt += f"CONTENT: {section['content']}\n"
    
    prompt += """
INSTRUCTIONS:
1. Maintain all the original headings and their levels
2. Preserve all the content and important information
3. Reformat everything to match the new template style
4. Do not add or remove substantive information
5. Format your response as valid Markdown with proper heading levels (# for level 1, ## for level 2, etc.)

REFORMATTED CONTENT:
"""
    return prompt
````

````python
def create_new_document(chunks_content: List[str], output_path: str) -> None:
    """
    Create a new document with the reformatted content.
    
    Args:
        chunks_content: List of strings containing reformatted content
        output_path: Path to save the new document
    """
    doc = Document()
    
    for chunk_content in chunks_content:
        # Parse markdown and add to document
        lines = chunk_content.strip().split('\n')
        
        for line in lines:
            if line.startswith('#'):
                # Count the number of # to determine heading level
                level = len(re.match(r'^(#+)', line).group(1))
                text = line.lstrip('#').strip()
                
                # Add heading to document
                heading = doc.add_heading(text, level=level)
            elif line.strip():
                # Add paragraph to document
                doc.add_paragraph(line)
    
    doc.save(output_path)
````

````python
### Using LLM

import docx
import json
import os
import boto3
from tqdm import tqdm
import argparse
from concurrent.futures import ThreadPoolExecutor

def extract_document_structure(doc_path):
    """Extract structure from a docx file."""
    print(f"Extracting structure from {doc_path}")
    doc = docx.Document(doc_path)
    structure = []
    
    current_header = None
    content_buffer = []
    
    for para in doc.paragraphs:
        if not para.text.strip():
            continue  # Skip empty paragraphs
        
        # Identify headers based on style
        if para.style.name.startswith("Heading 1") or para.style.name.lower() == "heading 1":
            # If we have a previous header with content, save it
            if current_header and content_buffer:
                structure.append({
                    "header": current_header,
                    "content": content_buffer.copy()
                })
            
            # Start new header section
            current_header = para.text
            content_buffer = []
        else:
            # Add paragraph to current content buffer
            content_buffer.append(para.text)
    
    # Add the last section if there is one
    if current_header and content_buffer:
        structure.append({
            "header": current_header,
            "content": content_buffer
        })
    
    print(f"Extracted {len(structure)} sections from {doc_path}")
    return structure

def chunk_document(structure, max_chunk_size=2000):
    """Break down document into smaller chunks within each header section."""
    chunks = []
    
    for section in structure:
        header = section["header"]
        content = section["content"]
        
        # Calculate total content length
        content_text = "\n".join(content)
        
        # If content is small enough, keep it as one chunk
        if len(content_text) <= max_chunk_size:
            chunks.append({
                "header": header,
                "content": content
            })
        else:
            # Split content into smaller chunks
            current_chunk = []
            current_size = 0
            
            for paragraph in content:
                para_size = len(paragraph)
                
                # If adding this paragraph exceeds the chunk size, save current chunk
                if current_size + para_size > max_chunk_size and current_chunk:
                    chunks.append({
                        "header": header,
                        "content": current_chunk.copy()
                    })
                    current_chunk = []
                    current_size = 0
                
                current_chunk.append(paragraph)
                current_size += para_size
            
            # Add the last chunk if there's anything left
            if current_chunk:
                chunks.append({
                    "header": header,
                    "content": current_chunk
                })
    
    print(f"Document broken into {len(chunks)} chunks")
    return chunks

def save_chunks_to_file(chunks, filename):
    """Save document chunks to a JSON file for checkpoint/resuming."""
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(chunks, f, indent=2, ensure_ascii=False)
    print(f"Chunks saved to {filename}")
    return filename

def load_chunks_from_file(filename):
    """Load document chunks from a JSON file."""
    with open(filename, 'r', encoding='utf-8') as f:
        chunks = json.load(f)
    print(f"Loaded {len(chunks)} chunks from {filename}")
    return chunks

def create_reformatting_prompt(chunk, template_info):
    """Create a prompt for Claude to reformat the content."""
    header = chunk["header"]
    content = chunk["content"]
    
    prompt = f"""Reformat the following document section according to this template:

Template information:
{json.dumps(template_info, indent=2)}

Document section to reformat:
Header: {header}

Content:
"""
    
    # Add content paragraphs
    for paragraph in content:
        prompt += f"{paragraph}\n\n"
    
    prompt += """
Please reformat this content following the template style.
Return ONLY the reformatted content in this JSON format:
{
  "header": "reformatted header text",
  "content": [
    "paragraph 1",
    "paragraph 2",
    ...
  ]
}
"""
    
    return prompt

def process_chunk_with_bedrock(chunk, template_info, bedrock_client, model_id):
    """Process a document chunk using Claude on AWS Bedrock."""
    # Create the prompt
    prompt = create_reformatting_prompt(chunk, template_info)
    
    # System prompt for Claude
    system_prompt = """You are a document formatting assistant. 
Your task is to reformat document sections according to the provided template.
Preserve all original content and meaning while improving the structure and clarity.
Return ONLY the JSON with the reformatted content."""
    
    # Create request body for Claude
    request_body = {
        "anthropic_version": "bedrock-2023-05-31",
        "max_tokens": 4000,
        "temperature": 0.1,
        "system": system_prompt,
        "messages": [
            {
                "role": "user",
                "content": prompt
            }
        ]
    }
    
    try:
        # Call Claude via Bedrock
        response = bedrock_client.invoke_model(
            modelId=model_id,
            body=json.dumps(request_body)
        )
        
        # Parse response
        response_body = json.loads(response.get("body").read())
        response_text = response_body["content"][0]["text"]
        
        # Extract JSON from response
        reformatted = extract_json_from_response(response_text)
        
        if reformatted:
            return reformatted
        else:
            print(f"Failed to parse response for chunk with header: {chunk['header'][:30]}...")
            return chunk  # Return original as fallback
            
    except Exception as e:
        print(f"Error processing chunk with Claude: {e}")
        return chunk  # Return original as fallback

def extract_json_from_response(response_text):
    """Extract JSON content from Claude's response."""
    import re
    
    # Try to find JSON block between triple backticks
    json_match = re.search(r'```(?:json)?\s*([\s\S]*?)\s*```', response_text)
    if json_match:
        json_str = json_match.group(1)
    else:
        # Try to find JSON object directly
        json_match = re.search(r'({[\s\S]*})', response_text)
        if json_match:
            json_str = json_match.group(1)
        else:
            json_str = response_text
    
    try:
        return json.loads(json_str)
    except json.JSONDecodeError:
        return None

def process_chunks_parallel(chunks, template_info, bedrock_client, model_id, max_workers=4):
    """Process chunks in parallel using ThreadPoolExecutor."""
    print(f"Processing {len(chunks)} chunks with Claude on Bedrock (max workers: {max_workers})")
    reformatted_chunks = []
    
    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = []
        for chunk in chunks:
            future = executor.submit(
                process_chunk_with_bedrock, 
                chunk, 
                template_info, 
                bedrock_client,
                model_id
            )
            futures.append(future)
        
        # Process results as they complete
        for i, future in enumerate(tqdm(futures, desc="Processing chunks")):
            try:
                result = future.result()
                reformatted_chunks.append(result)
            except Exception as e:
                print(f"Error processing chunk {i}: {e}")
                # Use original chunk as fallback
                reformatted_chunks.append(chunks[i])
    
    return reformatted_chunks

def create_new_document(reformatted_chunks, template_info, output_path):
    """Create a new document with the reformatted content."""
    print(f"Creating new document at {output_path}")
    
    # Create a new document
    doc = docx.Document()
    
    # Process each chunk
    for chunk in reformatted_chunks:
        # Add header
        header = chunk.get("header", "")
        if header:
            header_para = doc.add_paragraph(header)
            header_para.style = "Heading 1"
        
        # Add content paragraphs
        content = chunk.get("content", [])
        for paragraph in content:
            if paragraph.strip():
                para = doc.add_paragraph(paragraph)
                # Apply template style if specified
                if "paragraph_style" in template_info:
                    para.style = template_info["paragraph_style"]
    
    # Save the document
    doc.save(output_path)
    print(f"Document successfully saved to {output_path}")
    return output_path

def main():
    parser = argparse.ArgumentParser(description="Reformat document using Claude on AWS Bedrock")
    parser.add_argument("--input", required=True, help="Path to input document")
    parser.add_argument("--output", required=True, help="Path to output document")
    parser.add_argument("--aws-profile", help="AWS profile to use")
    parser.add_argument("--aws-region", default="us-east-1", help="AWS region for Bedrock")
    parser.add_argument("--model", default="anthropic.claude-3-sonnet-20240229-v1:0", 
                      help="Claude model ID in Bedrock")
    parser.add_argument("--chunks-file", help="Save/load chunks from this file")
    parser.add_argument("--max-workers", type=int, default=4, help="Maximum number of parallel workers")
    parser.add_argument("--chunk-size", type=int, default=2000, 
                      help="Maximum size of content chunks in characters")
    
    args = parser.parse_args()
    
    # Template information (could be loaded from a separate file)
    template_info = {
        "header_style": "Heading 1",
        "paragraph_style": "Normal",
        "formatting_rules": [
            "Make headers concise and descriptive",
            "Ensure paragraphs are well-structured",
            "Maintain original meaning and content",
            "Improve clarity and readability where possible"
        ]
    }
    
    # Set up AWS session and Bedrock client
    session_args = {}
    if args.aws_profile:
        session_args['profile_name'] = args.aws_profile
    
    session = boto3.Session(**session_args, region_name=args.aws_region)
    bedrock_client = session.client('bedrock-runtime')
    
    chunks = None
    
    # Load chunks from file if specified and file exists
    if args.chunks_file and os.path.exists(args.chunks_file):
        chunks = load_chunks_from_file(args.chunks_file)
    
    # Otherwise extract structure and create chunks
    if chunks is None:
        # Extract document structure
        structure = extract_document_structure(args.input)
        
        # Chunk the document
        chunks = chunk_document(structure, args.chunk_size)
        
        # Save chunks if a file was specified
        if args.chunks_file:
            save_chunks_to_file(chunks, args.chunks_file)
    
    # Process chunks with Claude
    reformatted_chunks = process_chunks_parallel(
        chunks,
        template_info,
        bedrock_client,
        args.model,
        args.max_workers
    )
    
    # Create new document
    output_path = create_new_document(reformatted_chunks, template_info, args.output)
    
    print(f"Document reformatting complete")
    print(f"Input: {args.input}")
    print(f"Output: {output_path}")
    print(f"Model used: {args.model}")
````
